```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Objective Function
def ObjectiveFunction(X):
    return X[0]**2 + X[1]**2  # Example: Sphere function for optimization

# Parameters
noP = 36  # Number of particles
nVar = 2  # Number of variables
lb = -10 * np.ones(nVar)  # Lower bound for particles
ub = 10 * np.ones(nVar)   # Upper bound for particles

# PSO Parameters
Max_iteration = 200
Vmax = 6
wMax = 0.9
wMin = 0.2
c1 = 2
c2 = 2

# Initialize swarm
x = np.linspace(-10, 10, 6)
y = np.linspace(-10, 10, 6)
X, Y = np.meshgrid(x, y)
particles = np.array([[X[i, j], Y[i, j]] for i in range(len(x)) for j in range(len(y))])

# Initializations
Swarm = {
    "Particles": [{"X": part, "V": np.random.rand(2)*Vmax, "PBEST": {"X": np.random.rand(2)*Vmax, "O": np.inf}} for part in particles],
    "GBEST": {"X": np.zeros(2), "O": np.inf}
}

# Setup for plot
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
x_vals = np.linspace(-10, 10, 30)
y_vals = np.linspace(-10, 10, 30)
X_new, Y_new = np.meshgrid(x_vals, y_vals)
Z = np.array([[ObjectiveFunction([X_new[i, j], Y_new[i, j]]) for j in range(len(x_vals))] for i in range(len(y_vals))])

contour = ax.contour(X_new, Y_new, Z, 20)
particles_plot, = ax.plot([], [], 'ok', markerfacecolor='k')
velocities_plot, = ax.plot([], [], '-k', lw=2.5)

# Function to update particle positions
def update(frame):
    global Swarm

    # Update the inertia weight
    w = wMax - frame * ((wMax - wMin) / Max_iteration)

    # Update the velocity and position of particles
    for k in range(noP):
        particle = Swarm["Particles"][k]
        particle["O"] = ObjectiveFunction(particle["X"])

        if particle["O"] < particle["PBEST"]["O"]:
            particle["PBEST"]["O"] = particle["O"]
            particle["PBEST"]["X"] = particle["X"]

        if particle["O"] < Swarm["GBEST"]["O"]:
            Swarm["GBEST"]["O"] = particle["O"]
            Swarm["GBEST"]["X"] = particle["X"]

        # Update velocity
        particle["V"] = w * particle["V"] + \
                        c1 * np.random.rand(2) * (particle["PBEST"]["X"] - particle["X"]) + \
                        c2 * np.random.rand(2) * (Swarm["GBEST"]["X"] - particle["X"])

        # Ensure velocity does not exceed limits
        particle["V"] = np.clip(particle["V"], -Vmax, Vmax)

        # Update position
        particle["X"] = particle["X"] + particle["V"]

        # Apply boundary conditions
        particle["X"] = np.clip(particle["X"], lb, ub)

    # Update plot
    x_data = np.array([p["X"][0] for p in Swarm["Particles"]])
    y_data = np.array([p["X"][1] for p in Swarm["Particles"]])

    particles_plot.set_data(x_data, y_data)

    if frame == 0:  # Only initialize velocities on first frame
        velocities_x = np.array([p["V"][0] for p in Swarm["Particles"]])
        velocities_y = np.array([p["V"][1] for p in Swarm["Particles"]])
        velocities_plot.set_data(velocities_x, velocities_y)

    return particles_plot, velocities_plot

# Create animation
ani = FuncAnimation(fig, update, frames=Max_iteration, interval=50, repeat=True)

plt.show()

# Plot convergence curve
cg_curve = [Swarm["GBEST"]["O"] for _ in range(Max_iteration)]
plt.figure()
plt.plot(cg_curve)
plt.xlabel('Iteration')
plt.ylabel('Objective Value')
plt.title('Convergence Curve')
plt.show()
```
